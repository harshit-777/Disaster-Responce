# 1. Import libraries and load datasets.
- Import Python libraries
- Load messages.csv into a dataframe and inspect the first few lines.
- Load categories.csv into a dataframe and inspect the first few lines.

```python
import pandas as pd 
import sqlalchemy as db
```

```python
messages_df = pd.read_csv('messages.csv',usecols=['id','message','original','genre'])
messages_df.head()
```

```python
messages_df.shape
```

```python
categories = pd.read_csv('categories.csv',usecols=['id','categories'])
```

```python
categories.head()
```

```python
categories.shape
```

# 2 Merge dataset

```python
df = pd.merge(messages_df,categories,on='id')
```

```python
df.shape
```

```python
df.head()
```

# 3 Split categories

```python
categories = df['categories'].str.split(';',expand=True)
```

```python
categories
```

```python
col_vals = categories.iloc[0].str.replace('-','')
```

```python
col_vals
```

```python
col_vals = col_vals.str.replace('0','')
col_vals = col_vals.str.replace('1','')
```

```python
col_vals
```

```python
categories.columns = list(col_vals)
```

```python
print(categories.columns)
```

```python
categories.head()
```

# Converting Categories into 0 ro 1

```python
for col in categories:
    categories[col] = categories[col].str[-1]
    categories[col] = categories[col].astype(int)
```

```python
categories.head()
```

# 4 Replace `categoris` column in `df` with new dataFrame

- drop categories column we dont need because we have simplified categories in simple way

```python
df = df.drop('categories',axis=1)
```

```python
df.head(3)
```

```python
df = pd.concat([df,categories],axis=1)
```
